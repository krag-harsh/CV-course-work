```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import imageio
```

```python
path="0002.jpg"
image = imageio.imread(path)  # reading the image
```

```python
plt.imshow(image)
plt.show()
```

```python
# printing shapes etc to get an idea
#(5,50)(5,100)(15,100)(7,100)(15,25)
sh=image.shape
print(sh)
im2d=image.reshape(-1,sh[2])  # reshaping the image to 2d array(from 3d)
print(im2d.shape)
```

```python
# clustering using DBSCAN
db = DBSCAN(eps=5, min_samples=50, metric = 'euclidean',algorithm ='auto').fit(im2d)
```

```python
# clustered = db.core_sample_indices_[db.labels_]
# print(clustered.shape)
la=db.labels_  # storing the labels
```

```python
plt.imshow(np.reshape(la, [sh[0], sh[1]]),cmap="gray")  # plotting the labels to get an idea of the clusters
```

Now changing the hyperparameters and plotting the images

```python
db2 = DBSCAN(eps=5, min_samples=100, metric = 'euclidean',algorithm ='auto').fit(im2d)
la2=db2.labels_
plt.imshow(np.reshape(la2, [sh[0], sh[1]]),cmap="gray")
# print("Number of clusters:",len(set(la2)))
```

```python
db3 = DBSCAN(eps=15, min_samples=100, metric = 'euclidean',algorithm ='auto').fit(im2d)
la3=db3.labels_
plt.imshow(np.reshape(la3, [sh[0], sh[1]]),cmap="gray")
# print("Number of clusters:",len(set(la3)))
```

```python
db4=DBSCAN(eps=7, min_samples=100, metric = 'euclidean',algorithm ='auto').fit(im2d)
la4=db4.labels_
plt.imshow(np.reshape(la4, [sh[0], sh[1]]),cmap="gray")
```
